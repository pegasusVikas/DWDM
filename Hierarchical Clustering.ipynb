```python
# import the required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Create the 200 sample data points
dataset = make_blobs(n_samples = 200,
                    centers = 4,
                    n_features = 2,
                    cluster_std = 1.6,
                    random_state = 50)
```

```python
# Gives points coordinates and clusters to which they belong
dataset
```

```python
# We require only coordinates
points = dataset[0]
points
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
# Create a dendrogram
dendrogram = sch.dendrogram(sch.linkage(points, method = 'ward'))
```

```python
plt.scatter(dataset[0][:,0],dataset[0][:,1])
plt.show()
```

```python
# Perform the actual clustering
hc = AgglomerativeClustering(n_clusters = 4,
                            affinity = 'euclidean',
                            linkage = 'ward')
```

```python
y_hc = hc.fit_predict(points)
```

```python
plt.scatter(points[y_hc == 0,0], points[y_hc == 0,1], s = 100, color = 'cyan')
plt.scatter(points[y_hc == 1,0], points[y_hc == 1,1], s = 100, color = 'yellow')
plt.scatter(points[y_hc == 2,0], points[y_hc == 2,1], s = 100, color = 'red')
plt.scatter(points[y_hc == 3,0], points[y_hc == 3,1], s = 100, color = 'green')

plt.show()
```
